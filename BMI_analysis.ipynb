```python
plt.figure(figsize = (18, 10))
(data.BMI.value_counts()).plot(kind='pie', autopct='%0.1f%%',startangle = 140)
plt.axis('equal')
plt.legend(title='CritÃ¨res', loc='upper left')
plt.title("RÃ©partition des CritÃ¨res du IMC", fontweight = 'bold')
```

```python
plt.figure(figsize = (18, 10))
(data.groupby('Gender')["BMI"].value_counts()).plot(kind='pie', autopct='%0.1f%%',startangle = 140)
plt.axis('equal')
plt.legend(title='CritÃ¨re par Sexe', loc='upper left')
plt.title("RÃ©partition des CritÃ¨res du IMC en Fonction du Sexe", fontweight = 'bold')
```
